```python
labels = ['Baseline','Baseline+Temp','Augmentation','Augmentation+Temp','Dropout','Dropout+Temp','Weight Decay','Weight Decay+Temp']
colors = ['darkgreen','lime','darkred','red','darkblue','lightblue','purple','violet']
```

```python
def plot_new(df,feature_column,feature_column_error,target_column,target_column_error):
    plt.figure(figsize=(8, 3))
    for i in range(len(df)):
        if i % 2 == 0:
            plt.errorbar(df[feature_column][i], df[target_column][i], xerr=df[feature_column_error][i], yerr=df[target_column_error][i], fmt='o', markersize=8, label=labels[i], color=colors[i],capsize=3)
        else:
            plt.errorbar(df[feature_column][i], df[target_column][i], xerr=df[feature_column_error][i], yerr=df[target_column_error][i], fmt='s', markersize=8, label=labels[i], color=colors[i],capsize=3)

    plt.xlabel(feature_column)
    plt.ylabel(target_column)
    plt.title(target_column + ' vs ' +feature_column)
    plt.legend()
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
    plt.savefig(f'scatter_plot_{feature_column}_{target_column}.pdf', bbox_inches='tight')
    plt.show()
```
